```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv('/Users/akanshamalviya/Documents/Input_Data/X_train_pca.csv')  # Replace 'your_dataset.csv' with the actual file path

# Display the shape of the dataset
data.shape
```

# Graph Showing Distribution Frequency of Label in the Training Dataset

```python
# Assuming 'label' is the column you want to visualize
label_counts = data['target'].value_counts()

# Plotting
plt.figure(figsize=(6, 4))
label_counts.plot(kind='bar', color='skyblue')
plt.title('Label Distribution')
plt.xlabel('target')
plt.ylabel('Frequency')
plt.show()
```

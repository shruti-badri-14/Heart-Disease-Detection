```python
import pandas as pd
```

### Load each of the four CSV files into separate DataFrames

```python
# Load each of the four CSV files into separate DataFrames
df1 = pd.read_csv('Processed CSVs/Processed_Data_Cleveland.csv')
df2 = pd.read_csv('Processed CSVs/Processed_Data_Hungary.csv')
df3 = pd.read_csv('Processed CSVs/Processed_Data_LongBeachVA.csv')
df4 = pd.read_csv('Processed CSVs/Processed_Data_Switzerland.csv')
```
